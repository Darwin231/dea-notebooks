```python
import datacube

dc = datacube.Datacube()

query_params = dict(x=(129.7, 129.8), 
                    y=(-14.75, -14.85), 
                    time=("1988", "2022"))

load_params = dict(dask_chunks={})

ds = dc.load(product="ga_ls8c_nbart_gm_cyear_3", 
             **query_params,
             **load_params)
```
