```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import math

def rmse(a, b):    
    try:
        return math.sqrt(mean_squared_error(a, b))
    except:
        df = pd.DataFrame({'a':a.values.flatten(), 'b': b.values.flatten()}).dropna()
        return math.sqrt(mean_squared_error(df.a, df.b))
    
def corr(a, b):    

    df = pd.DataFrame({'a':a.values.flatten(), 'b': b.values.flatten()}).dropna()
    return df.corr().iloc[0, 1]

def density_intersection(data1, data2, bins=1000, plot=False):
    
    hist_range = [min([data1.min(), data2.min()]),
                  max([data1.max(), data2.max()])]

    density1, bins1 = np.histogram(data1, bins=bins, range=hist_range)
    density2, bins2 = np.histogram(data2, bins=bins, range=hist_range)
    bins = (bins1[1:] + bins1[:-1]) / 2.0
    cumsum1 = np.cumsum(density1[::-1])[::-1] / density1.sum()
    cumsum2 = np.cumsum(density2) / density2.sum()

    diff = abs(cumsum2 - cumsum1)
    min_indices = np.where(diff == diff.min())[0]
    min_diff_x = bins[min_indices].mean()

    if plot:
        print(min_diff_x)
        plt.plot(bins, cumsum1)
        plt.plot(bins, cumsum2)
        plt.axvline(min_diff_x, color='red')
        
    return min_diff_x


def test_func(tide_m, class_i, intertidal):    

    if intertidal:
        data1 = tide_m[(class_i == 0)]
        data2 = tide_m[(class_i == 1)]
        return density_intersection(data1, data2, bins=1000, plot=False)
    else:
        return np.nan
```
